## 3. If-statements

### 3.1. If-statements with User-defined functions

```python
# Firstly, we will choose a subset of 1M rows, since the data is very large
df_subset = df_ords_prods_merged [:1000000]
```

```python
# Check dimensions
df_subset.shape
```

```python
# We define our new variable and give certain conditions 
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Now, to see the result of the function, we add a column to our df
df_subset['price_range'] = df_subset.apply(price_label, axis=1)
```

```python
# To check the values in our new column
df_subset['price_range'].value_counts(dropna = False)
```

```python
# There is no high range products so let's find out which product in the subset has the highest price
df_subset['prices'].max()
```

### 3.2. If-statements with the loc() function

```python
# Creating the conditions. For loc() function, we will be using one row per each function
df_subset.loc[(df_subset['prices'] <= 15) & (df_subset['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_subset.loc[df_subset ['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_subset.loc[df_subset['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_subset['price_range_loc'].value_counts(dropna=False)
```

```python
# Let's try to do it with our merged data
df_ords_prods_merged.loc[df_ords_prods_merged ['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices']> 15, 'price_range_loc'] = 'High range product'
```

```python
df_ords_prods_merged['price_range_loc'].value_counts(dropna = False)
```

### 3.3. If-statements with For-loops

```python
df_ords_prods_merged.head()
```

```python
# Creating a column to summarise how busy each day of the week is
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna=False)
```

#### Each number represent a day. From the project we know that Saturday = 0, meaning this is the busiest day of the week

```python
# Creating a column that displays the busiest, the least busy and regularly busy days of the weak
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

### 2. Create a "Busiest days" and "Less busy days" including two days in each list

```python
# Creating a column that displays the 2 busiest, the 2 less busy and regularly busy days of the week
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1 :
    result.append("Two_busiest_days")
  elif value == 3 or value == 4 :
    result.append("Two_slowest_days")
  else:
    result.append("Normal_days")
```

```python
result
```

```python
# To add the column to the dataframe
df_ords_prods_merged ['Busiest_days'] = result
```

### 3. Check the values in this new column

```python
df_ords_prods_merged ['Busiest_days'].value_counts(dropna = False)
```

#### We can see that Sat & Sun have almost as many orders as Mon, Thu & Fri together. Tue & Wed are the quietest days.

### 4. Create a new column for the busiest period of day

```python
# We create a column to summarise how busy each hour is
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
# Creating a column that displays the hours with the most orders, the least orders and the hours with average orders.
result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value == 3 or value == 4 or value == 2 or value == 5:
    result.append("Less busy hours")
  elif value == 1 or value ==0 or value ==6 or value ==23 or value ==22 or value ==21 or value ==7 or value ==20:
    result.append("Average busy hours")
  else:
    result.append("Most busy hours")
```

```python
# To add the column to the dataframe
df_ords_prods_merged ['busiest_period_of_day'] = result
```

```python
# To check the output
result
```

### 5. Check the frequency for this new column

```python
df_ords_prods_merged ['busiest_period_of_day'].value_counts (dropna = False)
```

#### Most orders occur between 10am-3pm. 
#### From 12 am-5am the amount of orders is quite small, thy will need less staff.
#### 5am-10am & 3pm-12am are considered periods with an average number of orders
